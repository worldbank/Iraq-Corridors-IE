```python
%matplotlib inline
import sys, os, logging

import rasterio

import osmnx as ox
import networkx as nx
import peartree as pt
import pandas as pd
import geopandas as gpd

from shapely.geometry import Point
```

```python
#Input AOI
inputAOI = r"Q:\WORKINGPROJECTS\CityScan\Data\CityExtents\Conotou_AOI.shp"
inAOI = gpd.read_file(inputAOI)
inAOI = inAOI.to_crs({'init': 'epsg:4326'})

#Input flood data
inFlood = r"Q:\GLOBAL\HYDRO\SSBN_Flooding\benin\BJ_pluvial_undefended\BJ-PU-100-1.tif"
inF = rasterio.open(inFlood)
```

```python
# Get the OSM graph for analysis
%time walkG = ox.graph_from_polygon(inAOI.geometry[0], network_type='walk')
```

```python
#plot the OSM graph
ox.plot_graph(walkG)
```

```python
#get geodataframe of osm nodes
osmNodes = [data for id, data in walkG.nodes(data=True)]

osmDF = pd.DataFrame(osmNodes)
```
