### Generate convex Hull - or 'Boundary' of GTFS Area
Could also be replaced by something else - e.g. urban boundary. This helper polygon is passed to the OSMNX function which pulls down the OSM as a graph in the next step

```python
import geopandas as gpd
from shapely.geometry import Point

# We need a coverage area, based on the points from the
# New Orleans GTFS data, which we can pull from the peartree
# network graph by utilizing coordinate values and extracting
# a convex hull from the point cloud
boundary = gpd.GeoSeries(
    [Point(n['x'], n['y']) for i, n in Ggtfs.nodes(data=True)]
    ).unary_union.convex_hull

# Visually inspect
boundary
```

### Pull in the walking and driving networks with OSMnx

```python
%time Gwalk = ox.graph_from_polygon(boundary, network_type='walk')
%time Gdrive = ox.graph_from_polygon(boundary, network_type='drive')
```

### Convert lengths in OSMnx graphs over to travel times for each segment
walking

```python
%%time

# Edge length in peartree is represented in seconds
# while in OSMnx it is in meters, so we need to convert those values
# down to seconds, based on some walk speed
walk_speed = 4.5  # about 3 miles per hour

# Make a copy of the graph in case we make a mistake
Gwalk_adj = Gwalk.copy()

# Iterate through and convert lengths to seconds
for from_node, to_node, edge in Gwalk_adj.edges(data=True):
    orig_len = edge['length']
    
    # Note that this is a MultiDiGraph so there could
    # be multiple indices here, I naively assume this is not
    # the case
    Gwalk_adj[from_node][to_node][0]['orig_length'] = orig_len

    # Conversion of walk speed and into seconds from meters
    kmph = (orig_len / 1000) / walk_speed
    in_seconds = kmph * 60 * 60
    Gwalk_adj[from_node][to_node][0]['length'] = in_seconds
    
    # And state the mode, too
    Gwalk_adj[from_node][to_node][0]['mode'] = 'walk'
```

Driving graph

```python
%%time

# Edge length in peartree is represented in seconds
# while in OSMnx it is in meters, so we need to convert those values
# to seconds
speed = {
    'residential': 20,  # kmph
    'primary': 40, # kmph
    'primary_link':35,
    'motorway':45,
    'motorway_link': 40,
    'trunk': 40,
    'trunk_link':35,
    'secondary': 30, # kmph
    'secondary_link':25,
    'tertiary':30,
    'tertiary_link': 25,
    'unclassified':20
}
    
# Make a copy of the graph in case we make a mistake
Gdrive_adj = Gdrive.copy()

# Iterate through and convert lengths to seconds
for from_node, to_node, edge in Gdrive_adj.edges(data=True):
    
    orig_len = edge['length']
    
    # Note that this is a MultiDiGraph so there could
    # be multiple indices here, I naively assume this is not
    # the case
    Gdrive_adj[from_node][to_node][0]['orig_length'] = orig_len

    # Conversion of walk speed and into seconds from meters
    # get appropriate speed limi
    highwayclass = Gdrive_adj[from_node][to_node][0]['highway']
    if type(highwayclass) == list:
        highwayclass = highwayclass[0]
    
    # perform conversion
    kmph = (orig_len / 1000) / speed[highwayclass]
    in_seconds = kmph * 60 * 60
    Gdrive_adj[from_node][to_node][0]['length'] = in_seconds
    
    # And state the mode, too
    Gdrive_adj[from_node][to_node][0]['mode'] = 'drive'
```

```python
# Ensure that we now have both length values (and
# thus an updated edge schema)
list(Gwalk_adj.edges(data=True))[0]
```

```python
# As you can see we are missing the
# boarding cost for the walk network nodes (which
# is just zero)
list(Gwalk_adj.nodes(data=True))[0]
```

```python
# So this should be easy - just go through all nodes
# and make them have a 0 cost to board
for i, node in Gwalk_adj.nodes(data=True):
    Gwalk_adj.node[i]['boarding_cost'] = 0

for i, node in Gdrive_adj.nodes(data=True):
    Gdrive_adj.node[i]['boarding_cost'] = 0
```

```python
# Now that we have a formatted walk network
# it should be easy to reload the peartree graph
# and stack it on the walk network
start = 7 * 60 * 60
end = 9 * 60 * 60

# Note this will be a little slow - an optimization here would be
# to have coalesced the walk network
%time Ggtfs = pt.load_feed_as_graph(feed, start, end, existing_graph=Gwalk_adj)
print('nodes Ggtfs: %d' % len(Ggtfs.nodes(data=True)))
print('edges Ggtfs: %d' % len(Ggtfs.edges(data=True)))

Gmaster = Ggtfs.copy()

# add on walking graph
Gmaster.add_nodes_from(Gwalk_adj)
Gmaster.add_edges_from(Gwalk_adj.edges())
print('nodes Gmaster: %d' % len(Gmaster.nodes(data=True)))
print('edges Gmaster: %d' % len(Gmaster.edges(data=True)))

# Add on driving graph
Gmaster.add_nodes_from(Gdrive_adj)
Gmaster.add_edges_from(Gdrive_adj.edges())
print('nodes Gmaster: %d' % len(Gmaster.nodes(data=True)))
print('edges Gmaster: %d' % len(Gmaster.edges(data=True)))
```

### nice! this block makes a node GeoDataFrame pretty smoothly...

```python
import pandas as pd
nodes = []

for i in list(Gwalk.nodes(data=True)):
    nodes.append({
        'node':i[0],
        'y':i[1]['y'],
        'x':i[1]['x'],
        'osmid':i[1]['osmid'],
        'geometry':Point((i[1]['x'], i[1]['y']))
        }
    )

nodes_df = pd.DataFrame(nodes)
nodes_df = nodes_df.drop_duplicates(subset=['node'], keep='first')
nodes_gdf = gpd.GeoDataFrame(nodes_df, geometry=nodes_df.geometry)
```

###  ...and from this I can plot it all out

```python
# Just a set of reds, from dark to light
colors = ['black', 'red']

reachable = nodes_gdf.loc[nodes_gdf.node.isin(test.nodes)]
not_reachable = nodes_gdf.loc[nodes_gdf.node.isin(test.nodes) == False]

gdf_bin = [not_reachable,reachable]

# Hold onto the axis state between plots
ax = None

# Note: Reversing order of both so that lightest is drawn first, others
#       are then stacked on top (as they get darker)

for gdf, col in zip((gdf_bin),(colors)):

    
    # Plot parameters
    a = 0.15
    lw = 0.25
    if ax is not None:
        gdf.plot(ax=ax, linewidth=lw, alpha=a, color=col, markersize=3)
    else:
        ax = gdf.plot(figsize=(14,14), linewidth=lw, alpha=a, color=col, markersize=3)
```
